```python
# crie um codigo que receba uma base e depois um expoente e calcule a potencia
base = float(input("digite a base: "))
expoente = float(input("digite o expoente: "))
potencia = base ** expoente
print(f"A potência de {base} elevado a {expoente} é {potencia}")
```

```python
# crie um programa que solicite a o usuario a quantidade de horas trabalhadas por semana e o valor da hora de trabalho e calcule o salario mensal
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas por semana: "))
valor_hora = float(input("Digite o valor da hora de trabalho: "))
salario_mensal = horas_trabalhadas * valor_hora
print(f"O salário mensal é: R${salario_mensal}")
```
